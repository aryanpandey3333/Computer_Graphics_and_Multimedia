# ***Implementing Translational, Rotational, Shearing, Scaling, Reflection for 2D***

```python
import cv2
from skimage import io
from google.colab.patches import cv2_imshow
import numpy as np

img = io.imread('https://thumbs.dreamstime.com/z/alien-spacecraft-appeared-over-ancient-cities-science-fiction-illustration-200252855.jpg')
img = cv2.cvtColor(img, cv2.COLOR_RGB2BGR)
cv2_imshow(img)

resized_img = cv2.resize(img, None, fx=0.5, fy=0.5, interpolation = cv2.INTER_CUBIC)
cv2_imshow(resized_img)

rows, cols, dim = resized_img.shape
center = (cols/2, rows/2)
rotate_matrix = cv2.getRotationMatrix2D(center = center, angle = 90, scale = 1)
rotated_img = cv2.warpAffine(src = resized_img, M = rotate_matrix, dsize = (cols, rows))
cv2_imshow(rotated_img)

M = np.float32([[1,0,100],[0,1,120]])
translated_img = cv2.warpAffine(resized_img,M,(cols,rows))
cv2_imshow(translated_img)

Mx = np.float32([[1, 0, 0], [0, -1, rows], [0, 0, 1]])
reflected_img_x = cv2.warpPerspective(resized_img, Mx, (int(cols),int(rows)))
cv2_imshow(reflected_img_x)

Mx = np.float32([[1, 0.5, 0], [0, 1, 0], [0, 0, 1]])
sheared_img_x = cv2.warpPerspective(resized_img , Mx, (int(cols*1.5), int(rows*1.5)))
cv2_imshow(sheared_img_x)
```
