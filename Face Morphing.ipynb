```python
import facemorpher
import os
import requests
```

```python
img_data = requests.get('https://www.plastikapraha.cz/data/2017/01/23/14/img-face-mobile.png').content
with open('dest.jpg', 'wb') as handler:
    handler.write(img_data)
img_data = requests.get('https://st.depositphotos.com/1008939/1880/i/950/depositphotos_18807295-stock-photo-portrait-of-handsome-man.jpg').content
with open('src.jpg', 'wb') as handler:
    handler.write(img_data)
```

```python
facemorpher.morpher(["src.jpg", "dest.jpg"], plot=True)
```
